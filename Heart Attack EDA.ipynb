```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\CUS\Desktop\EDA FILES\heart_attack_dataset.csv')
```

```python
df.drop_duplicates()
```

```python
df.isnull().sum()
```

```python
df['Gender']=df['Gender'].map({'Male':1,'Female':0})
df['Has Diabetes']=df['Has Diabetes'].map({'Yes':1,'No':0})
df['Smoking Status']=df['Smoking Status'].map({'Smoker':1,'Non_Smoker':0})
df['Chest Pain Type']=df['Chest Pain Type'].astype('category').cat.codes
```

```python
df.head(5)
```

```python
sns.set(style='whitegrid')
plt.figure(figsize=(10,8))
sns.histplot(df['Age'],kde=True,color='blue')
plt.title('Distribution of Age')
plt.xlable('Age')
plt.ylable('Frequancy')
plt.show()
```

```python
#blood Distribution
plt.figure(figsize=(10,8))
sns.histplot(df['Blood Pressure (mmHg)'],kde=True, color='red')
plt.title('Distribution of Blood Pressure')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.histplot(df['Cholesterol (mg/dL)'],kde=True, color='green')
plt.title('Distribution of Cholesterol (mg/dL)')
plt.show()
```

```python
#count Gender
plt.figure(figsize=(10,8))
sns.countplot(x='Gender',data=df , palette='pastel')
plt.title('Count of Gender')
plt.xticks(ticks=[0,1],lables=['Female','Male'])
plt.show()
```

```python
#Distribution of Cholesterol by Gender
plt.figure(figsize=(10,8))
sns.boxplot(x='Gender',y='Cholesterol (mg/dL)',data=df,palette='pastel')
plt.title('Cholesterol Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Cholesterol (mg/dL)')
plt.xticks(ticks=[0,1],labels=['Female','Male'])
plt.show()
```

```python
#chest pain type vs.Age 
plt.figure(figsize=(12,7))
sns.boxplot(x='Chest Pain Type',y='Age',data=df,palette='Set2')
plt.title('Chest Pain Type')
plt.ylabel('Age')
plt.show()
```

```python
# Count of Chest Pain Type
plt.figure(figsize=(10,6))
sns.countplot(x='Chest Pain Type',data=df,palette='viridis')
plt.xlabel('Chest Pain Type')
plt.ylabel('Count')
plt.show()
```
